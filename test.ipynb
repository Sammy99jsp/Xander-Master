```python
from typing import Generator

from xander import Game, Arena, Reaction, Turn, Agent, AttackOfOpportunity, Creature


def routine() -> Generator[None, Reaction | Turn | None, None]:
    while True:
        t = yield
        if t is None:
            break

        match t:
            case Turn() as turn:
                print("Got a turn!")
                turn.end()
                pass
            case Reaction(AttackOfOpportunity() as aoo):
                print("Got an AOO!")
                for attack in aoo.available_attacks():
                    print(attack.name, attack.type, attack.range)
                aoo.skip()
                pass
 
game = Game(Arena(30, 30))
game.join(Agent("Test", routine(), seed=0), Creature.test())
game.start()
```
